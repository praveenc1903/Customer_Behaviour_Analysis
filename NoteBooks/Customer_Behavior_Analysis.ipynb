```python
# Define dtype mapping
dtype_mapping = {
    'int64': 'INTEGER',
    'float64': 'FLOAT',
    'object': 'TEXT',
    'datetime64[ns]': 'DATETIME',
    'bool': 'BOOLEAN'
}
```

```python
dtype = [dtype_mapping[str(i)] for i in customer_journey.dtypes]
table_column = ",".join(f"{i} {j}" for i, j in zip(customer_journey.columns, dtype))
print(table_column)
```

```python
sql_schemas = {}
for table_name, df in df_dict.items():
    columns_sql = []
    for col, dtype in zip(df.columns, df.dtypes):
        sql_type = dtype_mapping.get(dtype, 'VARCHAR(255)')  # Default fallback
        columns_sql.append(f"{col} {sql_type}")

    create_table_sql = f"CREATE TABLE {table_name} (\n  " + ",\n  ".join(columns_sql) + "\n);"
    sql_schemas[table_name] = create_table_sql
    print(f"-- Schema for {table_name}")
    print(create_table_sql)
    print("\n")
```

### DATA INSERTION AUTOMATICALLY 

```python
import pandas as pd
import mysql.connector
from mysql.connector import Error

try:
    # Dictionary mapping tables to their CSV files
    table_files = {
        'customer_journey': '/Users/PRAVEEN/Project_1/customer_journey.csv',
        'customer_reviews': '/Users/PRAVEEN/Project_1/customer_reviews.csv',
        'customers': '/Users/PRAVEEN/Project_1/customers.csv',
        'engagement_data': '/Users/PRAVEEN/Project_1/engagement_data.csv',
        'geography': '/Users/PRAVEEN/Project_1/geography.csv',
        'products': '/Users/PRAVEEN/Project_1/products.csv'
    }

    if myconnection:  # Connection check that works with all versions
        cursor = myconnection.cursor()
        
        for table_name, file_path in table_files.items():
            try:
                # Read CSV file
                df = pd.read_csv(file_path)
                
                # Clean data: replace NaN with None for MySQL NULL
                df = df.where(pd.notnull(df), None)
                
                # Generate INSERT query
                columns = ', '.join(df.columns)
                placeholders = ', '.join(['%s'] * len(df.columns))
                query = f"INSERT INTO {table_name} ({columns}) VALUES ({placeholders})"
                
                # Convert DataFrame to list of tuples with proper NULL handling
                data_tuples = [tuple(None if pd.isna(value) else value for value in row) 
                              for row in df.to_numpy()]
                
                # Execute batch insert
                cursor.executemany(query, data_tuples)
                myconnection.commit()
                
                print(f"Successfully inserted {len(df)} rows into {table_name}")
                
            except Error as e:
                print(f"Error inserting into {table_name}: {e}")
                myconnection.rollback()
                continue
            except Exception as e:
                print(f"Unexpected error with {table_name}: {e}")
                myconnection.rollback()
                continue
                
except Error as e:
    print(f"Database connection error: {e}")
except Exception as e:
    print(f"General error: {e}")
'''finally:
    # Proper cleanup
    if 'myconnection' in locals() and myconnection:
        if 'cursor' in locals():
            cursor.close()
        myconnection.close()
        print("Database connection closed")'''
```
